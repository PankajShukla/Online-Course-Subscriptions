```python
def EDA(data, data_name, column_list=[]):
    
    print('\n')
    print(data_name)
    
    print('\nPrint first 3 records: ')
    display(data.head(3))
    
    print('\nPrint shape: ')
    print(data.shape)
    
    print('\nPrint null count: ')
    print(data.isnull().sum())
    
    print('\nPrint data type: ')
    print(data.dtypes)
    
    print('\nPrint data description type: ')
    print(data.describe())
    try:
        for col in column_list:
            
            print('\nPrint distribution (top 10) for field: ', col)
            print(data[col].value_counts().head(10))
            print('\n')
    except:
        pass
```

## 3.a. Order date and month format

## 3.c. Order - Subscription month derivation

```python
order['Line Items'].str.split(' ').head()
```

```python
order['subscription_type'] = np.where(order['Line Items'].str.lower().str.contains('month'), 'Monthly',
                                      np.where(order['Line Items'].str.lower().str.contains('year'), 'Yearly',
                                               'Others'))
```

```python
print(order[order.subscription_type=='Monthly']['Line Items'].unique())
print(order[order.subscription_type=='Yearly']['Line Items'].unique())
print(order[order.subscription_type=='Others']['Line Items'].unique())
```

```python
"""
--------------------------------------------------------------------------------------------------------
Multiplication Factor extraction from Line Item Text
--------------------------------------------------------------------------------------------------------
"""

multiplication_factor = []
for multivalues in range(100):
    multiplication_factor.append('x'+str(multivalues))
print(multiplication_factor[:10])
print(sorted(multiplication_factor, reverse=True)[:10])

def assign_multiplication_factor(row):
    for v in sorted(multiplication_factor, reverse=True):
        if row['Line Items'].find(v) > -1:
            return v
    
order['multiplication_factor'] = order.apply(assign_multiplication_factor, axis=1)

print(order['multiplication_factor'].unique())

print(order['subscription_type'].value_counts())

print(order[['subscription_type','multiplication_factor']].drop_duplicates())
```

```python
order['multiplication_factor'] = np.where(order['subscription_type']=='Yearly','x12', order['multiplication_factor'])
order['subscription_period_months'] = order['multiplication_factor'].str[1:].astype(int)
```

# 6. Creating a MRR calculating function

```python
def monthly_active_subscription_value(row, month_col, rev_col):
    if month_col>=row['Subscription_Start_Month'] and month_col<row['Subscription_End_Month']:
        return row[rev_col]
    else:
        return 0
    
    

    
def monthly_active_subscription_count(row, month_col, user_col):
    if month_col>=row['Subscription_Start_Month'] and month_col<row['Subscription_End_Month']:
        return row['Customer ID']
    else:
        return None
    
    

def calculate_MRR(_df, _revenue_col, level=None):
    
    SubscriptionYearMonth = list(_df.OrderCreated_YearMonth.unique())
    
    _df[SubscriptionYearMonth]=0
    
    for mo in SubscriptionYearMonth:
        _df[mo] = _df.apply(monthly_active_subscription_value, month_col=mo, rev_col=_revenue_col, axis=1)

    if level:
        df_MRR = _df.groupby(level)[SubscriptionYearMonth].sum().reset_index()
        mo = len(SubscriptionYearMonth)
        df_MRR['ARR'] = df_MRR[SubscriptionYearMonth].sum(axis=1)/mo*12
    
    else:   
        df_MRR = _df[SubscriptionYearMonth].sum().reset_index()
        df_MRR.columns = ['Year_Month', 'MRR']
    
    for mo in SubscriptionYearMonth:
        _df.rename(columns={mo: mo+ ' Revenue'}, inplace=True)
     
    return _df, df_MRR




def calculate_MAS(_df, _user_col, level=None):
    
    SubscriptionYearMonth = list(_df.OrderCreated_YearMonth.unique())

    _df[SubscriptionYearMonth]=None

    for mo in SubscriptionYearMonth:
        _df[mo] = _df.apply(monthly_active_subscription_count, month_col=mo, user_col=_user_col, axis=1)
        

    if level:
        df_MAS = _df.groupby(level)[SubscriptionYearMonth].nunique().reset_index()
    else:   
        df_MAS = _df[SubscriptionYearMonth].nunique().reset_index()
        df_MAS.columns = ['Year_Month', 'MAS']

        
    for mo in SubscriptionYearMonth:
        _df.rename(columns={mo: mo+ ' Subscription'}, inplace=True)
     
    return _df, df_MAS



def calculate_ARR(df_MRR):
    
    SubscriptionYearMonth = list(order.OrderCreated_YearMonth.unique())
    mo = len(SubscriptionYearMonth)
    ARR = df_MRR['MRR'].sum()/mo*12
    
    return ARR
```
